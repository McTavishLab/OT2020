```python
from opentree import OT
```

```python
## Read in the OTTids from file to get the tree
```

```python
fi = open("Mammal_TNRS.tsv").readlines()
```

```python
ott_ids = set()
for lin in fi[1:]:
    ott_id = lin.split('\t')[2]
    try:
        ott_ids.add(int(ott_id))
    except:
        pass

```

```python
len(ott_ids)
```

```python
1086292 in ott_ids
```

```python
call_record = OT.ws.tree_of_life_induced_subtree(ott_ids=ott_ids)
```

```python
call_record.response_dict
```

```python
assert('unknown' in call_record.response_dict), call_record.response_dict
unknown_ids = call_record.response_dict['unknown']
for u in unknown_ids:
    assert u.startswith('ott')
    ui = int(u[3:])
    if ott_ids and (ui in ott_ids):
            ott_ids.remove(ui)
```

```python
ui in ott_ids
```

```python
len(ott_ids)
```

```python
call_record = OT.ws.tree_of_life_induced_subtree(ott_ids=ott_ids)
```

```python
call_record.response_dict
```

```python
ott_node_info = {}

in_tree = set()
not_in_tree = set()

for ott_id in ott_ids:
    node_info = OT.synth_node_info(ott_id = ott_id)
    try:
        nid = node_info.response_dict.get('node_id')
        in_tree.add(ott_id)
    except:
        assert(str(node_info.response) == '<Response [400]>')
        nid=None
        not_in_tree.add(ott_id)
    ott_node_info [ott_id] =  nid
```

```python
output = OT.synth_induced_tree(ott_ids = list(ott_ids), label_format='name')
```

```python
treefile = "OT_vertlife_mammals.tre"
output.tree.write(path = treefile, schema = "newick")
```

```python
# Don't forget to cite your friendly phylogeneticists!
studies = output.response_dict['supporting_studies']
cites = OT.get_citations(studies) #this can be a bit slow
fi = open("OT_mammal_cites.txt","w")
fi.write(cites)
fi.close()
```
