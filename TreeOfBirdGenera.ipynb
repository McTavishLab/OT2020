```python
from opentree import OT, taxonomy_helpers, util
```

```python
# Download the OpenTree taxonomy
tax_path = taxonomy_helpers.download_taxonomy_file(loc = '..', version = "3.2")
```

```python
aves = OT.get_ottid_from_name('Aves')
```

```python
cleaned = taxonomy_helpers.clean_taxonomy_file(taxonomy_file="../ott3.2/taxonomy.tsv")
bird_genera = taxonomy_helpers.get_ott_ids_group_and_rank(group_ott_id=aves, rank='genus', taxonomy_file=cleaned)
```

```python
resp = OT.synth_induced_tree(ott_ids = bird_genera, label_format="id", ignore_unknown_ids=True)
```

## Doing some digging into this taxon, all descendents are extinct, so we don't have a good placement for it, and have dropped it

```python
tree=taxonomy_helpers.label_broken_taxa(synth_resp=resp, label_format='id')
```

```python
alltax = set(tips+nodes)
```
